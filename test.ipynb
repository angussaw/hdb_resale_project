```python
import pandas as pd
import os
import glob
```

```python
all_files = glob.glob(os.path.join("data/raw/for_training/" , "*.csv"))
```

```python
li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```
